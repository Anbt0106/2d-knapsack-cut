```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Đọc dữ liệu từ file CSV
df = pd.read_csv('experiment_results.csv')

sns.set(style='whitegrid')

# Profit
plt.figure(figsize=(10, 6))
plt.plot(df['n_items'], df['PTAS_profit'], marker='o', label='PTAS Profit')
plt.plot(df['n_items'], df['DP_profit'], marker='o', label='DP Profit')
plt.plot(df['n_items'], df['EDA_profit'], marker='o', label='EDA Profit')
plt.title('Profit Comparison')
plt.xlabel('Number of Items')
plt.ylabel('Profit')
plt.legend()
plt.tight_layout()
plt.show()

# thời gian thực thi
plt.figure(figsize=(10, 6))
plt.plot(df['n_items'], df['PTAS_time_ms'], marker='o', label='PTAS Time (ms)')
plt.plot(df['n_items'], df['DP_time_ms'], marker='o', label='DP Time (ms)')
plt.plot(df['n_items'], df['EDA_time_ms'], marker='o', label='EDA Time (ms)')
plt.title('Execution Time Comparison')
plt.xlabel('Number of Items')
plt.ylabel('Time (ms)')
plt.legend()
plt.tight_layout()
plt.show()

# bộ nhớ
plt.figure(figsize=(10, 6))
plt.plot(df['n_items'], df['PTAS_mem_B'], marker='o', label='PTAS Memory (B)')
plt.plot(df['n_items'], df['DP_mem_B'], marker='o', label='DP Memory (B)')
plt.plot(df['n_items'], df['EDA_mem_B'], marker='o', label='EDA Memory (B)')
plt.title('Memory Usage Comparison')
plt.xlabel('Number of Items')
plt.ylabel('Memory (Bytes)')
plt.legend()
plt.tight_layout()
plt.show()
```
